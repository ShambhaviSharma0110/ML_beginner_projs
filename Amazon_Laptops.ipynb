```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get('https://www.amazon.in/s?k=laptops&crid=2F9DW60WM98WD&qid=1675533839&sprefix=laptops%2Caps%2C226&ref=sr_pg_1',headers=headers).text
```

```python
soup = BeautifulSoup(webpage,'lxml')
```

```python
soup.find('span', class_ = 'a-size-medium a-color-base a-text-normal').text.strip()
```

```python
len(soup.find_all('div',class_ = 'sg-col-inner'))
```

```python
laptops = soup.find_all('div',class_ = 'sg-col-inner')
```

```python
model = []
ratings = []
reviews = []
price = []

for i in laptops:
  try:
   model.append(i.find('h2', class_ = 'a-size-mini a-spacing-none a-color-base s-line-clamp-2').text.strip())
  except:
    model.append(np.nan)
  try:
    ratings.append(i.find('a', class_ = 'a-popover-trigger a-declarative').text.strip())
  except:
    ratings.append(np.nan)
  try:
    reviews.append(i.find('span', class_ = 'a-size-base s-underline-text').text.strip())
  except:
    reviews.append(np.nan)
  try:
    price.append(i.find('span', class_ = 'a-price-whole').text.strip())
  except:
    price.append(np.nan)

  df = pd.DataFrame({
    'model': model,
    'ratings': ratings,
    'reviews': reviews,
    'price': price
})

  df.head()
```

```python
df.dropna()
```

**FOR MULTIPLE PAGES**

```python
for j in range(1,20):
  headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
  webpage=requests.get('https://www.amazon.in/s?k=laptops&crid=2F9DW60WM98WD&qid=1675533839&sprefix=laptops%2Caps%2C226&ref=sr_pg_{}'.format(j),headers=headers).text
  soup = BeautifulSoup(webpage, 'lxml')
  laptops = soup.find_all('div',class_ = 'sg-col-inner')
  model = []
ratings = []
reviews = []
price = []

for i in laptops:
  try:
   model.append(i.find('h2', class_ = 'a-size-mini a-spacing-none a-color-base s-line-clamp-2').text.strip())
  except:
    model.append(np.nan)
  try:
    ratings.append(i.find('a', class_ = 'a-popover-trigger a-declarative').text.strip())
  except:
    ratings.append(np.nan)
  try:
    reviews.append(i.find('span', class_ = 'a-size-base s-underline-text').text.strip())
  except:
    reviews.append(np.nan)
  try:
    price.append(i.find('span', class_ = 'a-price-whole').text.strip())
  except:
    price.append(np.nan)

  df1 = pd.DataFrame({
    'model': model,
    'ratings': ratings,
    'reviews': reviews,
    'price': price
})

  df1.head()
```
